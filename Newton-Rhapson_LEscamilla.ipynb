# Con iteracion

```python
import numpy as np
def f(x):
    return x**3-np.cos(x)
def df(x):
    return 3*x**2+np.sin(x)

x0=1
i=1
for iteracion in range(1,6):
    x1=x0-f(x0)/df(x0)
    x0=x1
    print (i, x1)
    i=i+1
    
print(x0)
```

# Con while

```python
def f(x):
    return x**3-np.cos(x)
def df(x):
    return 3*x**2+np.sin(x)

x0=20
i=1
error = 10
while error > 1e-6:
    x1=x0-f(x0)/df(x0)
    error = abs(x1-x0)
    x0=x1
    print (i, x1)
    i=i+1
    
print(x0)
```

# Con Bisect

```python
import scipy.optimize as optimize
import numpy as np

def func(x):
    return x**3-np.cos(x)

print(optimize.bisect(func, 0, 10))
```

```python
def samesign(a, b):
        return a * b > 0

def bisect(func, low, high):
    

    assert not samesign(func(low), func(high))

    for i in range(54):
        midpoint = (low + high) / 2.0
        if samesign(func(low), func(midpoint)):
            low = midpoint
        else:
            high = midpoint

    return midpoint

def f(x):
        return x**3-np.cos(x)

x = bisect(f, 0, 1)
print x, f(x)
```
